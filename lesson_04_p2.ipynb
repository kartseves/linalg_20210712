### __1.__ Решить систему уравнений методом Крамера:

###   а) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$
    
###   б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1
\end{cases}$

### __Решение:__

### а)

```python
import numpy as np

A = np.array([[1, -2], [3, -4]])
detA = np.linalg.det(A)
print(f'Матрица A:\n{A}')
print(f'Определитель матрицы A:\n{detA}')

A_1 = np.array([[1, -2], [7, -4]])
detA_1 = np.linalg.det(A_1)
print(f'Определитель матрицы A1:\n{detA_1}')
A_2 = np.array([[1, 1], [3, 7]])
detA_2 = np.linalg.det(A_2)
print(f'Определитель матрицы A2:\n{detA_2}')

x = np.array([[detA_1/detA, detA_2/detA]]).T
print(f'Вектор x:\n{x}')
print(f'Проверка, Ax:\n{A.dot(x)}=b')
```

```python
A = np.array([[2, -1, 5], [1, 1, -3], [2, 4, 1]])
detA = np.linalg.det(A)
print(f'Матрица A:\n{A}')
print(f'Определитель матрицы A:\n{detA}')

A_1 = np.array([[10, -1, 5], [-2, 1, -3], [1, 4, 1]])
detA_1 = np.linalg.det(A_1)
print(f'Определитель матрицы A1:\n{detA_1}')
A_2 = np.array([[2, 10, 5], [1, -2, -3], [2, 1, 1]])
detA_2 = np.linalg.det(A_2)
print(f'Определитель матрицы A2:\n{detA_2}')
A_3 = np.array([[2, -1, 10], [1, 1, -2], [2, 4, 1]])
detA_3 = np.linalg.det(A_3)
print(f'Определитель матрицы A3:\n{detA_3}')

x = np.array([[detA_1/detA, detA_2/detA, detA_3/detA]]).T
print(f'Вектор x:\n{x}')
print(f'Проверка, Ax:\n{A.dot(x)}=b')
```

### __Ответ:__ a) $x=\begin{pmatrix}
5 \\ 
2 
\end{pmatrix}$; б) $x=\begin{pmatrix}
2 \\ 
-1 \\
1
\end{pmatrix}$

### __5*.__ Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.

### __Решение:__

```python
def A_to_LU(a):
    lu = np.matrix(np.zeros([a.shape[0], a.shape[1]]))
    n = a.shape[0]
    for k in range(n):
        for j in range(k, n):
            lu[k, j] = a[k, j] - lu[k, :k] * lu[:k, j]
        for i in range(k + 1, n):
            lu[i, k] = (a[i, k] - lu[i, : k] * lu[: k, k]) / lu[k, k]
    return lu

def get_L(lu):
    L = lu.copy()
    for i in range(L.shape[0]):
        L[i, i] = 1
        L[i, i+1 :] = 0
    return L

def get_U(lu):
    U = lu.copy()
    for i in range(1, U.shape[0]):
        U[i, :i] = 0
    return U

def get_x(lu, b):
    y = np.matrix(np.zeros([lu.shape[0], 1]))
    for i in range(y.shape[0]):
        y[i, 0] = b[i, 0] - lu[i, :i] * y[:i]
    
    x = np.matrix(np.zeros([lu.shape[0], 1]))
    for i in range(1, x.shape[0] + 1):
        x[-i, 0] = (y[-i] - lu[-i, -i:] * x[-i:, 0] )/ lu[-i, -i]
    return x
```

```python
A = np.matrix(np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]]))
b = np.matrix(np.array([[531, -460, 193]]).T)
print(f'Матрица A:\n{A}')
print(f'Вектор b:\n{b}')

LU = A_to_LU(A)
L = get_L(LU)
U = get_U(LU)

print(f'Матрица LU:\n{LU}')
print(f'Матрица L:\n{L}')
print(f'Матрица U:\n{U}')
print(f'Матрица L*U:\n{L.dot(U)}=A')

print(f'Вектор х:\n{get_x(LU, b)}')
```
