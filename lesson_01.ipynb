### __1.__ Исследовать на линейную зависимость: $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$


### __Решение:__

Заметим, что:

### $$f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x),$$

то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}$ линейно зависимы.

### __Ответ:__ линейно зависимы

### __2.__ Исследовать на линейную зависимость: $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$

### __Решение:__

Заметим, что:

### $$f_{4}(x)=f_{3}(x) + 2 \cdot f_{2}(x) + \frac{1}{2} \cdot f_{1}(x),$$

то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$ линейно зависимы.

### __Ответ:__ линейно зависимы

### __3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

### __Решение:__

Т.к. явно не указано в каком базисе даны координаты вектора $x$, принием что они даны в стандартном базисе.

Стандартный базис линейного пространства $\mathbb{R}^{3}$ образуют векторы 
$e_{1}=(1, 0, 0)$, $e_{2}=(0, 1, 0)$, $e_{3}=(0, 0, 1)$.

$$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=2\cdot(1, 0, 0)+3\cdot(0, 1, 0)+5\cdot(0, 0, 1)=2e_{1}+3e_{2}+5e_{3},$$

тогда, так как:

$$e_{1}=\frac{1}{2} \cdot b_{2}$$

$$e_{2}= b_{3}$$

$$e_{3}=\frac{1}{10} \cdot b_{1}$$

то:

$$x=2 \cdot \frac{1}{2} \cdot b_{2} + 3 \cdot b_{3} + 5 \cdot \frac{1}{10} \cdot b_{1} = \frac{1}{2} \cdot b_{1} + b_{2} + 3 \cdot b_{3},$$

то есть координатами вектора $x$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$ являются $\frac{1}{2}$, $1$, $3$.

### __Ответ:__ ($\frac{1}{2}$, $1$, $3$)

### __4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

### а) в базисе $1$, $x$, $x^{2}$;

### б) в базисе $x^{2}$, $x-1$, $1$.

### __Решение:__

Примем что стандартный базис линейного пространства $\mathbb{R}^{3}[x]$ образуют векторы:

$$e_{1}=x^2=(1, 0, 0),$$

$$e_{2}=x=(0, 1, 0),$$

$$e_{3}=1=(0, 0, 1).$$

$$3x^{2}-2x+2=3e_{1}-2e_{2}+2e_{3},$$

то есть координаты в исходном базисе ($3$; $-2$; $2$).

### а)

Базис $1$, $x$, $x^{2}$ образуют вектора $b_{1}=1=(0, 0, 1)$, $b_{2}=x=(0, 1, 0)$, $b_{3}=x^{2}=(1, 0, 0)$.

Тогда:

$$e_{1}= b_{3}$$

$$e_{2}= b_{2}$$

$$e_{3}= b_{1}$$


$$3e_{1}-2e_{2}+2e_{3}=3b_{3}-2b_{2}+2b_{1}=2b_{1}-2b_{2}+3b_{3},$$

то есть координатами вектора в базисе $1$, $x$, $x^{2}$ являются $2$, $-2$, $3$.

### б)

Базис $x^{2}$, $x-1$, $1$ образуют вектора $b_{1}=x^{2}=(1, 0, 0)$, $b_{2}=x-1=(0, 1, -1)$, $b_{3}=1=(0, 0, 1)$.

Тогда:

$$e_{1}= b_{1}$$

$$e_{3}= b_{3}$$

$$e_{2}-e_{3}=b_{2} \Rightarrow e_{2}=b_{2}+e_{3}=b_{2}+b_{3}$$

$$3e_{1}-2e_{2}+2e_{3}=3b_{1}-2(b_{2}+b_{3})+2b_{3}=3b_{1}-2b_{2},$$

то есть координатами вектора в базисе $x^{2}$, $x-1$, $1$ являются $3$, $-2$, $0$.

### __Ответ:__ а) ($2$, $-2$, $3$);  б) ($3$, $-2$, $0$).

### __5.__ Установить, является ли линейным подпространством:

### а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;

### б) все векторы, являющиеся линейными комбинациями данных векторов  $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

### __Решение:__

### а)

Возьмем два вектора трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю:

### $$a_{1}=\{0, a, b\},$$
### $$a_{2}=\{с, 0, d\}$$

Проведя проверку, мы увидим, что 
 
### $$(0, a, b)+(c, 0, d)=(c, a, b+d),$$

У данного вектора по крайней мере одна из первых двух координат не всегда будет равна нулю, а значит данное множество не является подпространством линейного пространства $\mathbb{R}^{3}$.

### б)

Возьмем два вектора, являющиеся линейными комбинациями векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$:

### $$a_{1}=\lambda_{1}u_{1}+\lambda_{2}u_{2}+...+\lambda_{n}u_{n},$$
### $$a_{2}=\mu_{1}u_{1}+\mu_{2}u_{2}+...+\mu_{n}u_{n},$$

Проведя проверку, мы увидим, что 
 
### $$\lambda_{1}u_{1}+\lambda_{2}u_{2}+...+\lambda_{n}u_{n}+\mu_{1}u_{1}+\mu_{2}u_{2}+...+\mu_{n}u_{n}=(\lambda_{1}+\mu_{1})u_{1}+(\lambda_{2}+\mu_{2})u_{2}+...(\lambda_{n}+\mu_{n})u_{n},$$

### $$\alpha\cdot (\lambda_{1}u_{1}+\lambda_{2}u_{2}+...+\lambda_{n}u_{n})=\alpha\lambda_{1}u_{1}+\alpha\lambda_{2}u_{2}+...+\alpha\lambda_{n}u_{n},$$

Полученные векторы также принадлежат указанному в задании множеству всех векторов, то есть данное множество является подпространством заданного линейного пространства.

### __Ответ:__ а) не является;  б) является.
